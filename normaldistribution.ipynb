```python
Python script to plot a normal distribution curve given a mean and standard deviation.
```

```python
import numpy as np
import matplotlib.pyplot as plt 
from scipy.stats import norm

mean = 0
std_dev = 1

data = np.random.normal(mean, std_dev, 1000)
x = np.linspace(mean-4*std_dev, mean+4*std_dev, 1000)
pdf_values = norm.pdf(x, mean, std_dev)
plt.hist(data, bins=30, density=True, alpha=0.5, color='g', label='Random Data') 
plt.plot(x, pdf_values, 'r', label='Normal distribution pdf')
plt.xlabel('Value')
plt.ylabel('Density')
plt.title(f"Normal Distribution (mean(mean), std_dev=(std_dev))")
plt.legend()
plt.grid(True)
plt.show()
```
